### **Q . Plot the histograms for each of the columns in mtcars using hist() function with bins as 10 and figsize of (20,15) .**

```python
plt.figure(figsize = (15,35))
mtcars_numeric = mtcars.select_dtypes(include=['float64', 'int64'])

for i, col in enumerate(mtcars_numeric.columns):
    plt.subplot(6, 2, i + 1)
    plt.hist(mtcars_numeric[col], bins = 10)
    plt.title(f'Histogram of {col}', fontsize=18)
    plt.xlabel(col, fontsize=14)
    plt.ylabel('Frequency', fontsize=14)

plt.tight_layout()
plt.show()
```

### **Q . Plot the density curves for each column of mtcars using 'for' loop and plot command .**

```python
plt.figure(figsize = (12,12))
for i, col in enumerate (mtcars_numeric.columns):
    plt.subplot(4, 3, i + 1)
    mtcars_numeric[col].plot(kind = "density")
    plt.xlabel((col), fontsize = 12)
    plt.title(f"Density of {col}")
plt.tight_layout()
plt.show()
```

### **Q . Generate 10000 random normally distributed numbers and plot the density curve along with the median and mean.**

```python
norm_dist = pd.DataFrame(np.random.normal(size=10000))

plt.figure(figsize = (10, 6))
norm_dist.plot(kind = "density", figsize = (10,6),color = 'blue', title =('Density Curve of Normally Distributed Numbers'), legend = False)
mean_normal = norm_dist.mean()[0]
median_normal = norm_dist.median()[0]

plt.axvline(mean_normal, color = 'red', linestyle = '--', label = (f'Mean: {mean_normal:.2f}'))
plt.axvline(median_normal, color = 'green', linestyle = '--', label = (f'Median: {median_normal:.2f}'))
plt.xlim(-4,4)
plt.legend()
plt.grid(True)
plt.show()      
```

### **Q .  Generate 10000 exponentially distributed numbers using np.random.exponential(size=100000) and plot the density curve along with the median and mean.**

```python
exp_dist = pd.DataFrame(np.random.exponential(size = 100000))

plt.figure(figsize=(10,10))
exp_dist.plot(kind = "density", color = "blue", figsize = (10,10), title = "Density curve of Exponentially distributed Data", legend = False)
exp_mean = exp_dist.mean()[0]
exp_median = exp_dist.median()[0]

plt.axvline(exp_mean, color = 'red', linestyle = '--', label = (f'Mean: {exp_mean:.2f}'))
plt.axvline(exp_median, color = 'green', linestyle = '--', label = (f'Median: {exp_median:.2f}'))
plt.legend()
plt.grid(True)
plt.xlim(-1,5)
plt.show()
```
